```python
import pandas as pd

sume_aesa = pd.read_csv("data/sume_aesa.csv")
sume_aesa['year'] = sume_aesa['Data da Medição'].apply(lambda x: x.split('/')[-1])
sume_aesa['month'] = sume_aesa['Data da Medição'].apply(lambda x: x.split('/')[-2])
sume_aesa['Volume Útil (hm³)'] = sume_aesa['Volume Útil (hm³)'].apply(lambda x: float(x.replace(',', '.')) if isinstance(x, str) else x)
sume_aesa['Volume Útil (m_2)'] = sume_aesa['Volume Útil (hm³)'].apply(lambda x: x * 1000000000 / 10e6)
```

```python
plot_water_over_time(
    sume_aesa,
    y_variables=['Volume Útil (hm³)'],
    labels=["Volume Útil (m_2)"],
    title="Área de água ao longo do tempo - com remoção de nuvens"
)
```
