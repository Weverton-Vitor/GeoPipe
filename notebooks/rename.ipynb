```python
import os
import glob
import numpy as np
```

```python


# Caminho da raiz onde estão suas pastas e arquivos
root_path = "C:/Users/weverton.vitor/Documents/faculdade/pibic/fmask-pipeline/data/"
sufixo = "_deepwatermap_fmask"

# Pega todos os arquivos e pastas (recursivamente)
todos_os_caminhos = glob.glob(os.path.join(root_path, "**"), recursive=True)

# Ordena os caminhos do mais longo para o mais curto para renomear pastas depois dos arquivos
todos_os_caminhos.sort(key=lambda x: -len(x))
print(todos_os_caminhos)

for caminho_antigo in todos_os_caminhos:
    pasta_pai, nome = os.path.split(caminho_antigo)

    nome_novo = nome.replace(sufixo, "")  # Remove o sufixo
    caminho_novo = os.path.join(pasta_pai, nome_novo)

    print(f"Renomeando:\n  {caminho_antigo}\n  → {caminho_novo}")
    os.rename(caminho_antigo, caminho_novo)
```
